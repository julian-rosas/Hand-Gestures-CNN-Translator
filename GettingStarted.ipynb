# 🧠 Getting started with Sign Language Digit Recognition with Custom CNN (NumPy + Numba)

This notebook demonstrates how to run a Convolutional Neural Network (CNN) from scratch using NumPy and Numba to recognize digits in sign language. The model is trained and evaluated on the Sign Language MNIST dataset (digits 0–9 subset).

---

## 📁 Step 1: Setup and Imports

First, ensure the required Python files are in the same directory:
- `CNN.py`: Contains the CNN model implementation
- `main.py`: Contains data loading, training, and evaluation functions


Also, install required packages:
```bash
pip install -r requirements.txt
```

Now let's import everything

````python
from CNN import *
from main import *
import time
````

## 📊 Step 2: Load and Preprocess Data

The dataset should be located in:

        - ./input/train/sign_mnist_train.csv

        - ./input/test/sign_mnist_test.csv

````python
X_train, y_train, X_test, y_test = load_and_preprocess_data()
print("Train samples:", len(X_train))
print("Test samples:", len(X_test))
print("Image shape:", X_train[0].shape)
````

## 🏋️‍♂️ Step 3: Train the CNN Model

This trains the CNN with two convolutional layers and a fully connected output layer.

````python
start_time = time.time()

results = train_cnn_model(X_train, y_train, X_test, y_test)
training_costs, training_accuracies, trained_parameters, layer_dimensions, label_template = results

print("Training complete in", round(time.time() - start_time, 2), "seconds.")
````

## 📈 Step 4: Visualize Training Progress

````python
visualize_training_progress(training_costs, training_accuracies)
````

## 🧪 Step 5: Evaluate the Model

Run the trained model on the test dataset and check accuracy.

````python
accuracy = evaluate_model_accuracy(X_test, y_test, trained_parameters, layer_dimensions, label_template)
print(f"Test accuracy: {accuracy:.2f}%")
````

## ✅ Summary

This CNN was implemented entirely from scratch using:

-    NumPy for numerical ops

-    Numba to accelerate performance

-    No deep learning libraries like TensorFlow or PyTorch

You can tune parameters such as learning rate, batch size, and epochs in main.py.
